```python
# Importing important liberaries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading dataset
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
# Columns of dataset.
df.columns
```

```python
#shape of data-set
df.shape
```

```python
# Information about dataset
df.info()
```

```python
# Checking for duplication.
df.duplicated().sum()
```

```python
# Dropping duplicates value.
df.drop_duplicates()
```

```python
# dropping Country column.
df = df.drop(['Country'],axis=1)
```

```python
# dropping Postal Code column.
df = df.drop(['Postal Code'],axis=1)
```

```python
# Shape of data after dropping.
df.shape
```

```python
# Describing the dataset.
df.describe()
```

```python
# Plotting the pair plot to see the interdependent distribution.
plt.figure(figsize=(5,5))
sns.pairplot(df)
```

```python
# Plotting heatmap to see correlation visualy.
corr = df.corr()
plt.figure(figsize=(10,7))
sns.heatmap(corr,cmap='YlGnBu',annot=True)
```

```python
value = df['Category'].value_counts()
labels= df['Category'].value_counts().index
```

```python
# Plotting pie chart for 'Category'.
plt.figure(figsize=(5,5))
plt.title('Category')
plt.pie(value,labels=labels,autopct='%.05f%%',shadow = True,startangle = 90, textprops={'fontsize': 18})
plt.show()
```

```python
# Sorting the 'Profit' in the descending order.
df1 = df.sort_values(['Profit','Category'],ascending = False)
df1.head()
```

```python
# Plotting the  line graph for 'Sales vs. Profit' for different catogery.
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
sns.set(font_scale=1.5)
sns.lineplot(x='Sales',y='Profit',data=df1[ :1500], palette='rainbow',hue='Category',linewidth=2 )
plt.title("Profit Vs. Sales on the basis of Category")
```

## Conclusion:
#### (i)  We should increase the sales of Technology rather than furniture to maximise our profit becuase from the above pie chart ,we observed that sales of technology (18.48%)  is less than furniture(21.22%).



```python
# Plotting the  line graph for 'Profit vs. Discount' for different catogery.
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
sns.set(font_scale=1.5)
sns.lineplot(x='Discount',y='Profit',data=df1[ :1000], palette='spring',hue='Category',linewidth=2)
plt.title("Profit Vs. Discout on the basis of category")
```

### Conclusion :
#### (i) For the maximum profit on technology, we should not give discount more than 10%.
#### (ii) For the maximum profit on furniture , we should not give discount more than 10%.
#### (iii) for the maximum profit on office supplies, we should give discount more than 10% becuause it  is gradually increasing.

```python
# computing top sub-categories in terms of Sales from first 150 observations
top_subcategory_Sales = df.groupby("Sub-Category").Sales.sum().nlargest(n = 150)

# computing top sub-categories in terms of profit from first 150 observations
top_subcategory_Profit = df.groupby("Sub-Category").Profit.sum().nlargest(n = 150)
```

```python
# Polotting bar plot for 'Sub-Category' in terms of Sales and Profit
sns.set_style('darkgrid')
top_subcategory_Sales.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='green')
top_subcategory_Profit.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'yellow')
plt.xlabel('Sub-Category',fontsize = 15)
plt.ylabel('Total (Sales/Profits)',fontsize = 15)
plt.title("Top Sub-Category (Sales vs Profit)",fontsize = 15)
plt.legend()
plt.show()
```

## Conclusion:
#### (i) Profit on the sales of table is negative . So we should decrease the sale on tables.
#### (ii) Profit on the sales of Copier and Phones are higher as compare to others 

```python
#Pottig pie chart for regional distribrution of Sales.
plt.figure(figsize=(5,5))
plt.title('Region',fontsize =15)
sns.set(palette='spring')
plt.pie(df['Region'].value_counts(),labels=df['Region'].value_counts().index,autopct='%.05f%%',shadow = True,startangle = 90, textprops={'fontsize': 15})
plt.show()
```

```python
# computing top sub-categories in terms of Sales from first 100 observations
top_regional_Sales = df.groupby("Region").Sales.sum().nlargest(n=100)

# computing top sub-categories in terms of profit from first 100 observations
top_regional_Profit = df.groupby("Region").Profit.sum().nlargest(n=100)
```

```python
# Plotting the bar graph for 'Region' in term of Sales & Profit.
sns.set_style('darkgrid')
top_regional_Sales.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='blue')
top_regional_Profit.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('Region',fontsize = 15)
plt.ylabel('Total (Sales/Profits)',fontsize = 15)
plt.title("Regional (Sales vs Profit)",fontsize = 15)
plt.legend()
plt.show()
```

## Conclusion:
#### As we can see that Profit in West and East is almost same in top 100 Sales & Profit even the Sales of East is less than West. So, we should increase the Sales in East.

```python
top_Statewise_Sales = df.groupby("State").Sales.sum().nlargest(n=10)
# computing top sub-categories in terms of profit from first 100 observations
top_Statewise_Profit = df.groupby("State").Profit.sum().nlargest(n=10)
```

```python
sns.set_style('darkgrid')
top_Statewise_Sales.plot(kind = 'bar',figsize = (10,5),fontsize = 14,color='purple')
top_Statewise_Profit.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'blue')
plt.xlabel('State',fontsize = 15)
plt.ylabel('Total (Sales/Profits)',fontsize = 15)
plt.title("Statewise (Sales vs Profit)",fontsize = 15)
plt.legend()
plt.show()
```

```python
# Evaluting the inter=dependecy of Profit, Sales and Discounts.
plt.figure(figsize=(10,5))
sns.set(font_scale=1.5)
sns.relplot(x="Sales",y="Profit",kind="scatter",size="Discount",hue="Discount",data=df,palette='rainbow')
```
